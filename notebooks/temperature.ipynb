### I. Check for temperature values and missing values in original MAC Season 4 Data
* Queried from betydb

```python
df = pd.read_csv('data/raw/mac_season_4.csv')
# df.head()
```

#### A. Drop columns

```python
cols_to_drop = ['Unnamed: 0', 'checked', 'result_type', 'citation_id', 'treatment_id', 'city', 'scientificname', 
                'commonname', 'genus', 'species_id', 'author', 'citation_year', 'treatment', 'month', 'year', 'dateloc', 
                'n', 'statname', 'stat', 'notes', 'entity', 'view_url', 'edit_url']

df_1 = df.drop(labels=cols_to_drop, axis=1)
# df_1.head()
```

#### B. Isolate only the rows with temperature-related traits

```python
temp_traits = [ ]

for trait in df.trait.unique():
    
    if 'temperature' in trait:
        print(trait)
        temp_traits.append(trait)
        
# print(temp_traits)
```

```python
df_2 = df_1.loc[df.trait.isin(temp_traits)]
```
