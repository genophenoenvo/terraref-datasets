````python
test_dates = dec_29.start_time.values

def just_dates(original_date_array):
    
    just_dates_list = []
    
    for date in original_date_array:
        
        date = date[5:10]
        just_dates_list.append(date)
            
    return just_dates_list
````

````python
just_dates(test_dates)
````

````python
start_dates = weather_2017.start_time.values
end_dates = weather_2017.end_time.values

new_start_dates = just_dates(start_dates)
````

````python
new_end_dates = just_dates(end_dates)
````

````python
weather_2017['short_start_date'] = new_start_dates
weather_2017['short_end_date'] = new_end_dates
````

````python
weather_2017.sample(n=20)
````

````python
# sample looks accurate so will drop the original date rows

new_dates_df = weather_2017.drop(labels=['start_time', 'end_time'], axis=1)
new_dates_df.head()
````
