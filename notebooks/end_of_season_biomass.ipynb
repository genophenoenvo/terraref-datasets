### End of season biomass
trait = `aboveground_dry_biomass`

```python
import datetime
import pandas as pd
import numpy as np
import sqlite3

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_0 = pd.read_csv('../data/raw/mac_season_4.csv')
df_0.head()
```

```python
# df_0.columns
```

```python
cols_to_keep = ['result_type', 'id', 'site_id', 'sitename', 'lat', 'lon', 'cultivar_id', 'author', 
                'date', 'time', 'raw_date', 'trait', 'trait_description', 'mean', 'units', 'notes', 
                'access_level', 'cultivar', 'entity', 'method_name']
```

```python
cols_to_drop = []

for col in df_0.columns.values:
    
    if col not in cols_to_keep:
        cols_to_drop.append(col)

cols_to_drop
```

```python
df_1 = df_0.drop(labels=cols_to_drop, axis=1)
df_1.head()
```

#### Work only with `aboveground_dry_biomass`

```python
df_1.trait.unique()
```

```python
df_2 = df_1.loc[df_1.trait == 'aboveground_dry_biomass']
print(df_2.shape)
# df_2.head()
```

```python
for col in df_2.columns:
    print(f'Number of unique values for {col}: {df_2[col].nunique()}')
```

```python
for col in df_2.columns:
    if df_2[col].nunique() < 5:
        print(f'Unique values for {col}: {df_2[col].unique()}')
```

```python
more_cols_to_drop = ['result_type', 'author', 'time', 'trait_description', 'notes', 'entity', 'method_name']

df_3 = df_2.drop(labels=more_cols_to_drop, axis=1)
# df_3.head()
```
